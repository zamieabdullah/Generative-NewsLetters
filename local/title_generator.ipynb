```python
from openai import OpenAI
import random
import os
from dotenv import load_dotenv
import json
import ast
```

```python
def get_topics(file_path):
    with open(file_path, 'r') as file:
        topic_list = []
        for line in file:
            topic_list.append(line.strip().lower())

    return topic_list

print(len(get_topics('../example_docs/seo_topics.txt')))
```

```python
load_dotenv()
gpt = os.getenv('gpt_token')
org = os.getenv('gpt_org')
client = OpenAI(api_key=gpt, organization=org)

key_topics = get_topics('../example_docs/seo_topics.txt')
master_seo_topics = []
master_prompt = """
    You are a real estate blog title creator. You will create real estate titles for the users and they will have to follow SEO guidelines.
    You are only going to return an arraylist of titles depending on how much the user asks for. Default to 1 if no size is given.
    Please remember what titles and topics you have written as you will be providing unique titles each time, depending on what the user asks for.
    Do not number them, put them within an arraylist.
    """
count = 1
for topics in key_topics:
    user_prompt = f"""
        can you provide 10 unique real estate topics or titles using this keyword: {topics}. I only need titles nothing else. 
        Please ensure that these topics are SEO titles as well.
        """

    response = client.chat.completions.create(
                    model="gpt-4o",
                    messages=[
                        {"role": "system", "content": master_prompt},
                        {"role": "user", "content": user_prompt},
                    ],
                    max_tokens=2000,
                    temperature=1.0,
    )

    try:
        actual_list = ast.literal_eval(response.choices[0].message.content)
        print("responded", count)
    except (ValueError, SyntaxError) as e:
        print("Error:", e)
    master_seo_topics = master_seo_topics + actual_list
    count += 1

print(len(master_seo_topics))
```

```python
with open('../example_docs/seo_key_topics.txt', 'w') as file:
    for item in master_seo_topics:
        file.write(f"{item}\n")
```
