```python
from openai import OpenAI
import random
import os
from dotenv import load_dotenv
import json
```

```python
def rank_seo(file_path):
    with open(file_path, 'r') as file:
        seo_dict = {}
        for line in file:
            term = line.strip().lower()
            if term not in seo_dict: 
                seo_dict[term] = 1
            else:
                seo_dict[term] += 1
    sorted_seo_dict = dict(sorted(seo_dict.items(), key=lambda item: item[1], reverse=True))
    return sorted_seo_dict


seo_dct = rank_seo("../example_docs/seo_tokens.txt")
print(seo_dct)
```

## For terms seen in only one document, these are seo terms that are seen frequently in that singular blog, but that also tells us that it was used for a very niche topic in real estate.

```python
seo_topics = {key: value for key, value in seo_dct.items() if value == 1}
print(list(seo_topics.keys()))
```

```python
with open('../example_docs/seo_topics.txt', 'w') as file:
    for item in seo_topics:
        file.write(f"{item}\n")
```
