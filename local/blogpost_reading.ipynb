# BiggerPockets Scraping

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.biggerpockets.com/blog/2024-dscr-loan-rates-and-how-to-track"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
article = soup.find_all('article')
article = article[0]

print(soup.find('h1').text)
print(soup.find('main').text)
```

```python
blogs = []
with open('../example_docs/biggerpockets.txt', 'r') as file:
    for line in file:
        url = line.strip()
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        article = soup.find_all('article')
        if len(article) > 0:
            article = article[0]
            title = article.find('h1').text.strip()
            content = article.find('main').text
            blogs.append({"title": title, "content": content})

print(len(blogs))
```

# Zillow Scraping

```python
url = "https://www.trulia.com/blog/most-affordable-neighborhoods-in-the-us/"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
print(soup.find('h1').text.strip())
print(soup.find('main').text.strip())
```

```python
with open('../example_docs/trulia.txt', 'r') as file:
    for line in file:
        url = line.strip()
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        title = soup.find('h1').text.strip()
        context = soup.find('main').text.strip()
        blogs.append({"title": title, "context": context})
        
```

```python
print(len(blogs))
```
